```python
to_replace = {'A': 'А',
              'B': 'В',
              'E': 'Е',
              'K': 'К',
              'M': 'М',
              'H': 'Н',
              'O': 'О',
              'P': 'Р',
              'C': 'С',
              'T': 'Т',
              'X': 'Х',
              'Y': 'У'
              }
```

```python
import cv2
from skimage import io
```

```python
import os
data_path_in = 'E:/train_2'
data_path_out = 'E:/train_2_resized'
```

```python
test = os.listdir(data_path_in)[1:]
test
```

```python
for item in test:
    print(item)
    src = io.imread(data_path_in+'/{}'.format(item))
    
    # dsize
    dsize = (1200, 750)

    # resize image
    output = cv2.resize(src, dsize)
    
    string=item
    for key in to_replace.keys():
        string = string.upper().replace(to_replace[key],key)

    cv2.imwrite(data_path_out+'/{}'.format(string),cv2.cvtColor(output, cv2.COLOR_RGB2BGR)) 
```
