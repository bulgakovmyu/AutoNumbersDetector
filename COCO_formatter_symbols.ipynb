```python
N=0
images = []
categories = [{"supercategory": "none", "name": "O", "id": 0},
             {"supercategory": "none", "name": "1", "id": 1},
             {"supercategory": "none", "name": "2", "id": 2},
             {"supercategory": "none", "name": "3", "id": 3},
             {"supercategory": "none", "name": "4", "id": 4},
             {"supercategory": "none", "name": "5", "id": 5},
             {"supercategory": "none", "name": "6", "id": 6},
             {"supercategory": "none", "name": "7", "id": 7},
             {"supercategory": "none", "name": "8", "id": 8},
             {"supercategory": "none", "name": "9", "id": 9},
             {"supercategory": "none", "name": "A", "id": 10},
             {"supercategory": "none", "name": "B", "id": 11},
             {"supercategory": "none", "name": "C", "id": 12},
             {"supercategory": "none", "name": "E", "id": 13},
             {"supercategory": "none", "name": "H", "id": 14},
             {"supercategory": "none", "name": "K", "id": 15},
             {"supercategory": "none", "name": "M", "id": 16},
             {"supercategory": "none", "name": "O", "id": 17},
             {"supercategory": "none", "name": "P", "id": 18},
             {"supercategory": "none", "name": "T", "id": 19},
             {"supercategory": "none", "name": "X", "id": 20},
             {"supercategory": "none", "name": "Y", "id": 21}]
annotations = []
for i, el in enumerate(annotation_list):
    images.append({
                  "file_name": (el['filename']).split('\\')[-1],
                  "height": 80,
                  "width": 150,
                  "id": i+1
                    })
    for j, annot in enumerate(el['annotation']):
        annotations.append({  "id": (j+1)+i*10,
                              "bbox": [
                                annot['x']-annot['width']/2,
                                annot['y']-annot['height']/2,
                                annot['width'],
                                annot['height']
                              ],
                              "image_id": i+1,
                              "segmentation": [],
                              "ignore": 0,
                              "area": annot['width']*annot['height'],
                              "iscrowd": 0,
                              "category_id": int(annot['symbol'])})


coco_dict = {
              "type": "instances",
              "images": images,
              "categories": categories,
              "annotations": annotations
            }
```
