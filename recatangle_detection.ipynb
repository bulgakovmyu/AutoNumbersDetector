```python
import os
import xmltodict, json
import xml.etree.ElementTree as ET
```

```python
data_path = 'data\\train_notations'
data_path_out = 'data\\train'
```

```python
test = os.listdir(data_path)
```

```python
def parseXmlToJson(xml):
  response = {}

  for child in list(xml):
    if len(list(child)) > 0:
      response[child.tag] = parseXmlToJson(child)
    else:
      response[child.tag] = child.text or ''

    # one-liner equivalent
    # response[child.tag] = parseXmlToJson(child) if len(list(child)) > 0 else child.text or ''

  return response
```

```python
to_replace = {'A': 'А',
              'B': 'В',
              'E': 'Е',
              'K': 'К',
              'M': 'М',
              'H': 'Н',
              'O': 'О',
              'P': 'Р',
              'C': 'С',
              'T': 'Т',
              'X': 'Х',
              'Y': 'У'
              }
```

```python
annotations = []
for item in test:
    annotation = {}
    with open(os.path.join(data_path, item), 'r', encoding='utf-8') as fd:
        doc = xmltodict.parse(fd.read())
        string = item.split('.')[0]
        for key in to_replace.keys():
            string = string.upper().replace(to_replace[key],key)
        string = string.replace('-', '')
#         print(item, doc['annotation']['size'])
        annotation["annotation"] = {
                                        "class": "rect",
                                        "width": float(doc['annotation']['object']['bndbox']['xmax'])-float(doc['annotation']['object']['bndbox']['xmin']),
                                        "height": float(doc['annotation']['object']['bndbox']['ymax'])-float(doc['annotation']['object']['bndbox']['ymin']),
                                        "x": float(doc['annotation']['object']['bndbox']['xmin']),
                                        "y": float(doc['annotation']['object']['bndbox']['ymin']),
                                        "number": string}
    annotation["class"]= "image"
    annotation["filename"]= os.path.join(data_path, item)
    annotations.append(annotation)
```

```python
import json
with open(os.path.join(data_path_out, 'autonumber_annotations.json'), "w", encoding="utf-8") as file:
    json.dump(annotations, file)
```
